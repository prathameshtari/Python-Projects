```python
 import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import seaborn as sns
import statistics

air = pd.read_excel("AirQuality.xlsx", sep = ',')
air.head(15)
```

```python
air.info()
```

```python
# //Checking if there are null values in the iteration
boolean=air.notnull() 
# print(boolean)
# //If boolean contains null values 'all' will return False else True
all(boolean)
# // all gives True so there are no NAN values
```

'Notnull' function will check all the values in the iteration and if there are missing values it will be False and if there are no missing values it will be True. Then we can compare all the values with 'all' function which works similar to 'And' operator and will give True only if all the values in the iteration are True.

```python
# // The missing values in this set is replaced with -200 so making those values zero
air=air.replace(-200, 0)
print(air.head(15))
```

This dataset contains missing values but all the missing values were filled as '-200'. So we replace those missing values by zero.

```python
area = np.pi*3
plt.title('Scatter plot CO and PT08.S1(CO) ')
plt.xlabel('CO')
plt.ylabel('PT08.S1(CO)')
plt.scatter(air['CO'],air['PT08.S1(CO)'], s=area, alpha=0.5)
plt.show()
```

From the scatter plot we can see that there is a strong correlation between the CO and PT08.S1(CO). It is a positive correlation and if one increases then the other will increase too.

```python
area = np.pi*3
plt.title('Scatter plot NMHC(GT) and PT08.S2(NMHC) ')
plt.xlabel('NMHC(GT)')
plt.ylabel('PT08.S2(NMHC)')
plt.scatter(air['NMHC(GT)'],air['PT08.S2(NMHC)'], s=area, alpha=0.5)
plt.show()
```

From the scatter plot we can see that there is a strong correlation between the true hourly averaged overall Non Metanic HydroCarbons concentration in microg/m^3 and PT08.S2 (titania) hourly averaged sensor response (nominally NMHC targeted). It is a positive correlation and if one increases then the other will increase too.

```python
area = np.pi*3
plt.title('Scatter plot NOx(GT) and PT08.S3(NOx) ')
plt.xlabel('NOx(GT)')
plt.ylabel('PT08.S3(NOx)')
plt.scatter(air['NOx(GT)'],air['PT08.S3(NOx)'], s=area, alpha=0.5)
plt.show()
```

There is a strong negative correlation between true hourly averaged NOx concentration in ppb and PT08.S3 (tungsten oxide) hourly averaged sensor response (nominally NOx targeted),

```python
area = np.pi*3
plt.title('Scatter plot NO2(GT) and PT08.S4(NO2) ')
plt.xlabel('NO2(GT)')
plt.ylabel('PT08.S4(NO2)')
plt.scatter(air['NO2(GT)'],air['PT08.S4(NO2)'], s=area, alpha=0.5)
plt.show()
```

As we can see from the scatterplot there is little to no correlation between true hourly averaged NO2 concentration in microg/m^3 and PT08.S4 (tungsten oxide) hourly averaged sensor response (nominally NO2 targeted).

```python
area = np.pi*3
plt.title('Scatter plot C6H6(GT) and PT08.S5(O3) ')
plt.xlabel('C6H6(GT)')
plt.ylabel('PT08.S5(O3)')
plt.scatter(air['C6H6(GT)'],air['PT08.S5(O3)'], s=area, alpha=0.5)
plt.show()
```
